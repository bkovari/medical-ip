```python
def load_scan(path, extension = '.jpg'):
    file_list = []
    for dirName, subdirList, fileList in os.walk(path):
        for filename in fileList:
            if extension in filename.lower():
                file_list.append(os.path.join(dirName,filename))
    return file_list

def load_images(path, extension = '.jpg'):
    file_list = load_scan(path, extension)
    images = []
    for file in file_list:
        images.append(cv.imread(file))
    return images
```
